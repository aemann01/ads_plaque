```r
library(webr)
library(ggplot2)
```

```r
tax_count <- read.delim("tax_count_arc.txt")
head(tax_count)
PieDonut(tax_count, aes("Tooth_type", "Genus", count="count"), showRatioThreshold = F)
pdf("tax_arc_donut.pdf")
PieDonut(tax_count, aes("Tooth_type", "Genus", count="count"), showRatioThreshold = F)

dev.off()
```

```r
ec_count <- read.delim("ec_count_arc.txt")
head(ec_count)
PieDonut(ec_count, aes("Tooth_type", "EC", count="Count"), showRatioThreshold = F)
pdf("ec_arc_donut.pdf")
PieDonut(ec_count, aes("Tooth_type", "EC", count="Count"), showRatioThreshold = F)
dev.off()
```
